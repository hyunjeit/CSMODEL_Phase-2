## Dataset Description

### Short Description  
* ##### The Family Income and Expenditure Survey (FIES) 2012 from the Philippines Statistics Authority (PSA) is a nationally-representative, public use dataset of 40,171 sampled Filipino households. It records expenditures and other disbursements, housing characteristics, key demographics, income and other sources of earnings, and entrepreneural activities of the households during 2012.

### Dataset Collection Method  
* ##### **Design** - Three-stage, stratified cluster sampling based on the 2003 Master Sample. The 17 administrative regions served as the main strata, with barangays or groups of 500+ households used as the primary sampling units.  

* ##### **Fieldwork** - Conducted through two face-to-face interviews. One in July 2012 and another in January 2013, each capturing the household's income and expenditure for the preceding six months.  

* ##### **Respondents & Weighting** - Data was provided by the household head or another informed adult. The survey includes adjusted household weights to reflect selection probability and non-responses.

### Implications of the Data Collection Method on this Project's Insights  
* ##### **Recall Bias**: Because data is self-reported and uses a six-month recall period, there's potential for inaccurate reporting, especially for irregular income or expense sources.  
* ##### **Seasonality Captured, but Shocks Missed**: Two six-month snapshots help reflect seasonal trends but might miss short-term economic shocks.  
* ##### **Sampling Design Considerations**: The use of clustered sampling and weighting requires analysts to apply survey-weighted techniques to produce unbiased estimates.  
* ##### **Limited Scope**: Institutional populations (e.g., dorms, nursing homes) are excluded, so insights are only applicable to private households.

### Structure of the Data  
* ##### **Unit of Analysis**: Each row represents a family or household, as defined by the Philippine Statistics Authority (PSA).  
* ##### **Number of Observations**: 40,171 households.  
* ##### **Format**: Tabular data with 119 variables/columns.  
* ##### **Features**: Columns include household identifiers, demographics, income types, food and non-food expenditure categories, dwelling and utilities data, durable goods ownership, and welfare indicators.

### Question 1: What is the distribution of total household income?


We examine the distribution of total household income to establish a baseline for understanding income disparities and identifying which factors may influence higher earnings.

We first summarize the data of the total household income `['TOINC']` column. Particularly, we calculate the mean, median, mode, and standard deviation.

```python
mean = fies_df['TOINC'].mean()
median = fies_df['TOINC'].median()
mode = fies_df['TOINC'].mode()
std = fies_df['TOINC'].std()
print("Total Household Income Summary:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print("Mode:", mode.tolist())
print(f"Standard Deviation: {std:.2f}")
```

We will use a histogram to display show the distribution of our data. To determine the number of bins, we take the square root of the number of observations. This is the simplest and most common method.

```python
numBins = round(math.sqrt(fies_df['TOINC'].count()))
```

We set the x-axis limits from 0 up to only the 95th percentile value of the column so that the histogram would not be stretched out by outliers.

```python
plt.figure(figsize=(10, 6))
plt.hist(fies_df['TOINC'], bins=numBins, color='yellow', edgecolor='black')
plt.title('Distribution of Total Household Income')
plt.xlabel('Total Income (TOINC)')
plt.ylabel('Number of Households')
plt.xlim(0, fies_df['TOINC'].quantile(0.95))
plt.grid(True)
x_min = 0
x_max = fies_df['TOINC'].quantile(0.95)
step = 50000
xticks = np.arange(x_min, x_max + step, step)
plt.xticks(xticks)
plt.show()
```

The histogram clearly shows that there is a right (positive) skew, wherein the mean is greater than the median. This means that most people earn between Php 50,000 to Php 140,000. The average is bumped up by the higher income earners, earning more than Php 500,000.

We can also use pearson skewness to check for the skewness quantitatively.

```python
pearson_skew = 3 * (mean - median) / std
print(f"Pearson's Second Skewness: {pearson_skew:.2f}")
```

A value of 0.92 suggest that there is a strong right skew.

## Data Mining

For the data mining portion of this study, **association rule mining** was employed, as the majority of the selected variables were already categorical. This technique was used to uncover patterns and relationships between the independent variables and the dependent variable (**total income**).

### Further Data Preprocessing (Binning and One-hot encoding)

Prior to performing association rule mining, the dataset underwent further preprocessing to group continuous numerical variables into discrete categorical bins, specifically for age, total income, household size, and the number of household members engaged in paid/profit work.

```python
df = fies_df.copy()

region_map = {
    1:  "Region I - Ilocos Region",
    2:  "Region II - Cagayan Valley",
    3:  "Region III - Central Luzon",
    41: "Region IVa - Calabarzon",
    42: "Region IVb - Mimaropa",
    5:  "Region V - Bicol Region",
    6:  "Region VI - Western Visayas",
    7:  "Region VII - Central Visayas",
    8:  "Region VIII - Eastern Visayas",
    9:  "Region IX - Western Mindanao",
    10: "Region X - Northern Mindanao",
    11: "Region XI - Southern Mindanao",
    12: "Region XII - Central Mindanao",
    13: "Region XIII - NCR",
    14: "Region XIV - CAR",
    15: "Region XV - ARMM",
    16: "Region XVI - CARAGA"
}
df["REGION"] = df["W_REGN"].map(region_map).fillna("Other").astype("category")

df["SETTLEMENT"] = df["URB"].map({1: "Urban", 2: "Rural"}).astype("category")

df["SEX"] = df["SEX"]

df["MARITAL_STATUS"] = df["MS"]

age_bins = [15, 24, 34, 44, 54, 64, float('inf')]
age_labels = ['15-24', '25-34', '35-44', '45-54', '55-64', '65+']
df["AGE_GROUP"] = pd.cut(df["AGE"], bins=age_bins, labels=age_labels, right=True).astype("category")

df["EDUCATION"] = df["HGC"]

df["JOB"] = df["JOB"]

df["INCOME_BRACKET"] = df["TOINC"].apply(lambda inc: (
    "Poor"                        if inc <= 12_030 * 12 else
    "Low Income (not poor)"      if 12_030 * 12 < inc <= 24_060 * 12 else
    "Low Middle Income"          if 24_060 * 12 < inc <= 48_120 * 12 else
    "Middle Middle Class"        if 48_120 * 12 < inc <= 84_210 * 12 else
    "Upper Middle Income"        if 84_210 * 12 < inc <= 144_360 * 12 else
    "Upper Income (not rich)"    if 144_360 * 12 < inc <= 240_600 * 12 else
    "Rich"
)).astype("category")

df["HOUSEHOLD_SIZE"] = df["MEMBERS"].apply(lambda m: (
    "Small Household"    if m <= 3 else
    "Medium Household"  if 4 <= m <= 6 else
    "Large Household"
)).astype("category")

df["PAID_WORKER_SIZE"] = df["EMPLOYED_PAY"].apply(lambda x: (
    "No Paid Workers"          if x == 0 else
    "Small No. Paid Workers"   if x <= 3 else
    "Medium No. Paid Workers"  if 4 <= x <= 6 else
    "Large No. Paid Workers"
)).astype("category")

df["PROFIT_WORKER_SIZE"] = df["EMPLOYED_PROF"].apply(lambda x: (
    "No Profit Workers"          if x == 0 else
    "Small No. Profit Workers"   if x <= 3 else
    "Medium No. Profit Workers"  if 4 <= x <= 6 else
    "Large No. Profit Workers"
)).astype("category")

df["SPOUSE_EMPLOYED"] = df["SPOUSE_EMP"]

binned_cols = [
    "REGION", "SETTLEMENT", "INCOME_BRACKET", "SEX",
    "AGE_GROUP", "MARITAL_STATUS", "EDUCATION", "JOB", "HOUSEHOLD_SIZE",
    "PAID_WORKER_SIZE", "PROFIT_WORKER_SIZE", "SPOUSE_EMPLOYED"
]

fies_fe = df[binned_cols]
```

First, a rule miner object must be instantianted with a given support and confidence threshold. For the support we decided to use **4%** of the number of observations in the dataset, and a confidence of **60%**. These values were not chosen arbitrarily but after testing different values. An initial value of 10% for support was chosen but was lowered for a richer rule set, and a confidence of 60% was chosen since the confidence values of the support values we tested were ranging from 20% to 80%, with a fair amount of values residing in 60% to 70%, so we decided with 60% as to not filter too much rules.

```python
rule_miner = RuleMiner(int(len(encoded_df) * 0.04) , 0.6)
```

Note that the function `get_association_rules()` returns only filtered rules where the consequent only contains income bracket variables to avoid unnecessary rules (e.g. IF AGE_GROUP_65+ AND SEX_MALE THEN SETTLEMENT_Rural)

```python
income_rules = rule_miner.get_association_rules(encoded_df)
```

```python
for rule in income_rules:
    X, y = rule
    confidence = rule_miner.get_confidence(encoded_df, [X, y])
    antecedent = " + ".join(X) if isinstance(X, list) else str(X)
    consequent = str(y)
    print(f"IF {antecedent} → THEN {consequent} | Confidence: {confidence:.2f}")
```

This suggests that poverty persists even among working individuals when the household lacks diversified income sources, spousal employment is absent, and the family resides in a rural setting.

The strongest rule, with a confidence of 77%, implies that even married men with jobs in rural areas, if living in households without paid workers, relying on small profit-generating activities, and having a non-working spouse, are very likely to be classified as poor.

This rule set reveals that employment alone does not guarantee an escape from poverty, especially when economic responsibility falls on a single earner and household productivity remains low.

## Statistical Inference

### Independent t-test (Welch's)

To conduct hypothesis testing between two different groups, we will be using an independent t-test. We use Welch's t-test by setting `equal_var=False` since the total household income `TOINC` data violates the assumption of equal variances.

### Household mean income by sex of household head

The summary statistics indicate that there is a positive skew for the total household income of both male and female lead households. This aligns with the positive skew we have observed earlier in the histogram for all households. Since the data is sufficiently large, there is no need to perform a log transformation as per the Central Limit Theorem.

```python
m_toinc = fies_df[fies_df["SEX"] == "Male"]["TOINC"]
f_toinc = fies_df[fies_df["SEX"] == "Female"]["TOINC"]

print("Male Income Summary Statistics")
print(m_toinc.agg(["mean", "std", "median"]))
print("Female Income Summary Statistics")
print(f_toinc.agg(["mean", "std", "median"]))

t_stat, p_value = stats.ttest_ind(m_toinc, f_toinc, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3e}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant income difference between male and female lead households.")
else:
    print("Fail to reject H0: There is no significant income difference between male and female lead households.")
```

### Household mean income by urbanity

Similar to the previous comparison, income data for both urban and rural households is positively skewed. Since the population is large, there is no need for normalization.

```python
u_toinc = fies_df[fies_df["URB"] == 1]["TOINC"]
r_toinc = fies_df[fies_df["URB"] == 2]["TOINC"]

print("Urban Income Summary Statistics")
print(u_toinc.agg(["mean", "std", "median"]))
print("Rural Income Summary Statistics")
print(r_toinc.agg(["mean", "std", "median"]))

t_stat, p_value = stats.ttest_ind(r_toinc, u_toinc, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3e}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant income difference between households living in urban and rural areas.")
else:
    print("Fail to reject H0: There is no significant income difference between households living in urban and rural areas.")
```

## ANOVA

To determine if there is a significant difference between the total household income across different groups, one-way ANOVA is used. This test compares the means of more than two independent groups, which can help determine if at least one group has a different average income as compared to the others. Since the sample size is large, the test is reliable even if the data is a bit skewed.

### Household mean income across different marital status of household heads

To determine if there is a significant difference in total household income across household head's marital status, the data was grouped by `MS` (Marital Status). Summary statistics were calculated for each group and used as input for a one-way ANOVA test using SciPy.

```python
ms_label = []
ms_toinc = []
for ms, group in fies_df.groupby("MS"):
    ms_label.append(ms)
    ms_toinc.append(group["TOINC"])

for i in range(len(ms_label)):
    print(ms_label[i])
    print(ms_toinc[i].agg(["mean", "std", "median"]))
    print("=========================================")

f_stat, p_value = stats.f_oneway(*ms_toinc)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.3e}")

if p_value < 0.05:
    print("Reject H0: There is a significant income difference between marital status groups.")
else:
    print("Fail to reject H0: No significant income difference between marital status groups.")
```

### Household mean income across different education level of household heads

To determine if there is a significant difference in total household income across household head's education level, the data was grouped by `HGC` (Highest Grade Completed). Summary statistics were calculated for each group and used as input for a one-way ANOVA test using SciPy.

```python
edu_label = []
edu_toinc = []
for edu, group in fies_df.groupby("HGC"):
    edu_label.append(edu)
    edu_toinc.append(group["TOINC"])

for i in range(len(edu_label)):
    print(edu_label[i])
    print(edu_toinc[i].agg(["mean", "std", "median"]))
    print("=========================================")

f_stat, p_value = stats.f_oneway(*edu_toinc)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.3e}")

if p_value < 0.05:
    print("Reject H0: There is a significant income difference between education groups.")
else:
    print("Fail to reject H0: No significant income difference between education groups.")
```

### Household mean income across different household head's age groups

To determine if there is a significant difference in total household income across household head's age groups, the data was grouped by `AGE_GROUP`. Summary statistics were calculated for each group and used as input for a one-way ANOVA test using SciPy.

```python
age_label = []
age_toinc = []
for age, group in fies_df.groupby("AGE_GROUP"):
    age_label.append(age)
    age_toinc.append(group["TOINC"])

for i in range(len(age_label)):
    print(age_label[i])
    print(age_toinc[i].agg(["mean", "std", "median"]))
    print("=========================================")

f_stat, p_value = stats.f_oneway(*age_toinc)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.3e}")

if p_value < 0.05:
    print("Reject H0: There is a significant income difference between age groups.")
else:
    print("Fail to reject H0: No significant income difference between age groups.")
```

## Spearman Correlation


### Correlation of Household head's age to Total Household Income

Spearman correlation will be used to analyze if there is a relationship between the household head's age and the total household income. This is used since spearman is less prone to outliers and does not require the data to follow a normal distribution.

```python
correlation, p_value = stats.spearmanr(fies_df["AGE"], fies_df["TOINC"])

print(f"Spearman Correlation between AGE and TOINC: {correlation:.3f}")
print(f"P-value: {p_value:.3e}")

alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant correlation between AGE and TOINC.")
else:
    print("Fail to reject H0: No significant correlation between AGE and TOINC.")
```

The Spearman correlation between AGE and total household income (TOINC) is 0.043 with a p-value less than 0.001, indicating a statistically significant but very weak positive correlation.
